```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
query1 = " SELECT * FROM heroes_information limit 10"
```

```python
pysqldf(query1)
```

```python
query2 = "SELECT DISTINCT  Publisher FROM heroes_information"
pysqldf(query2)
```

```python
query3 = "SELECT COUNT(name) FROM heroes_information WHERE Height > 200"
pysqldf(query3)
```

```python
query4 = "SELECT COUNT(name) FROM heroes_information WHERE Height > 200 AND Race = 'Human'"
pysqldf(query4)
```

```python
query5 = "SELECT COUNT(name) FROM heroes_information WHERE Weight BETWEEN 100 AND 200"
pysqldf(query5)
```

```python
query6 = "SELECT COUNT(name) FROM heroes_information WHERE \"Eye color\" = 'red' OR \"Eye color\" = 'blue'"
pysqldf(query6)
```

```python
query7 = "SELECT COUNT(*) FROM heroes_information WHERE (Race = 'Human' AND Race = 'Mutant') AND (\"Hair color\" = 'Green') OR (Race = 'Vampire' AND \"Hair color\" = 'Black')"
pysqldf(query7)
```

```python
query8 = "SELECT name FROM heroes_information ORDER BY Race  DESC"
pysqldf(query8)
```

```python
query9 = "SELECT COUNT(name) from heroes_information GROUP BY GENDER"
pysqldf(query9)
```

```python
query10 = "SELECT COUNT( DISTINCT Publisher) FROM heroes_information GROUP BY Publisher HAVING COUNT(*) > 15"
pysqldf(query10)
```
