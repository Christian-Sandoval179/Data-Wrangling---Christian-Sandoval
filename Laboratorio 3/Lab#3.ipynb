```python
query3 = """
    WITH inner_table AS (
        SELECT
            emp_no,
            salary,
            ROW_NUMBER() OVER (PARTITION BY emp_no ORDER BY salary DESC) AS seq
        FROM employee.salaries
    )
    SELECT
        B.first_name,
        B.last_name,
        B.birth_date,
        B.gender,
        A.emp_no,
        A.salary
    FROM inner_table A
    LEFT JOIN employee.employees B ON B.emp_no = A.emp_no
    WHERE A.seq = 1
    LIMIT 20
"""
pd.read_sql(query3, connection)
```
